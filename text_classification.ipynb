```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers,models
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.layers import LSTM,Dense,Input,Embedding,GlobalMaxPooling1D
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
dataa=pd.read_csv("spam.csv",encoding="ISO-8859-1")
```

```python
dataa.head
```

```python
dataa
```

```python
df=dataa.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1)
```

```python
df
```

```python
x=df["v2"]
x.shape
```

```python
df["yooo"]=df["v1"].map({"ham":0,"spam":1})
```

```python
y=df["yooo"].values
y
```

```python
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_test[0:5]
```

```python
no_words=2000
tokenizer = Tokenizer(num_words=no_words)
tokenizer.fit_on_texts(x_train)
sequence_train=tokenizer.texts_to_sequences(x_train)
sequence_test=tokenizer.texts_to_sequences(x_test)
```

```python
x_train=pad_sequences(sequence_train)

t=x_train.shape[1]

windx= tokenizer.word_index
v=len(windx)


print(x_train.shape)
```

```python
d=20
i=Input(shape=(t,))
x=Embedding(v+1,d)(i)
x=LSTM(15,return_sequences=True)(x)
x=GlobalMaxPooling1D()(x)
x=Dense(1,activation="sigmoid")(x)


model=Model(i,x)

model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train, num_classes=2)
y_test = to_categorical(y_test, num_classes=2)
```
